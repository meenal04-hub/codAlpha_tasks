**Split the data**

```python
X_train = train_data.drop('is_fraud', axis=1)  # Features for training
y_train = train_data['is_fraud']  # Target variable for training
X_test = test_data.drop('is_fraud', axis=1)  # Features for testing
y_test = test_data['is_fraud']  # Target variable for testing
```

```python
# Assuming 'X_train' and 'X_test' are your DataFrames
# Import LabelEncoder
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder from sklearn.preprocessing

for column in X_train.select_dtypes(include=['object']).columns:
    # Convert 'trans_date_trans_time' to numerical features
    if column == 'trans_date_trans_time':
        # Extract date and time components
        X_train['trans_date'] = pd.to_datetime(X_train[column]).dt.date
        X_train['trans_time'] = pd.to_datetime(X_train[column]).dt.time
        X_test['trans_date'] = pd.to_datetime(X_test[column]).dt.date
        X_test['trans_time'] = pd.to_datetime(X_test[column]).dt.time

        # Drop the original 'trans_date_trans_time' column
        X_train = X_train.drop(columns=[column])
        X_test = X_test.drop(columns=[column])

        # Convert date and time to numerical representations (e.g., ordinal)
        # For date: You can use toordinal()
        X_train['trans_date'] = X_train['trans_date'].apply(lambda date: date.toordinal())
        X_test['trans_date'] = X_test['trans_date'].apply(lambda date: date.toordinal())

        # For time: You can convert to total seconds since midnight
        X_train['trans_time'] = X_train['trans_time'].apply(lambda time: time.hour * 3600 + time.minute * 60 + time.second)
        X_test['trans_time'] = X_test['trans_time'].apply(lambda time: time.hour * 3600 + time.minute * 60 + time.second)

    else:
        # Create a LabelEncoder object
        label_encoder = LabelEncoder()

        # Get all unique values from both training and testing data
        all_values = list(set(X_train[column].unique()) | set(X_test[column].unique()))

        # Fit the encoder on all unique values from both training and testing data
        label_encoder.fit(all_values)

        # Transform both training and testing data for the current column
        X_train[column] = label_encoder.transform(X_train[column])
        X_test[column] = label_encoder.transform(X_test[column])
```

**Train and evaluate models**



1.   Logistic Regression




Strengths: Simple to implement, interpretable, and efficient for binary classification tasks like fraud detection. Works well when the relationship between features and target variable is linear.

Weaknesses: May not capture complex non-linear relationships in the data. Can be sensitive to outliers.

```python
# 1. Logistic Regression
lr_model = LogisticRegression(max_iter=1000)  # Increased max_iter to ensure convergence
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
```

Evaluate the models:

Calculate metrics like accuracy, precision, recall, and F1-score to compare the performance of the models.

```python
# Evaluate models
def evaluate_model(predictions, model_name):
    print(f"--- {model_name} ---")
    print("Accuracy:", accuracy_score(y_test, predictions))
    print("Precision:", precision_score(y_test, predictions))
    print("Recall:", recall_score(y_test, predictions))
    print("F1-Score:", f1_score(y_test, predictions))
    print("\n")
```

```python
evaluate_model(lr_predictions, "Logistic Regression")
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_metrics(metrics, model_name):
    """
    Creates a bar plot to visualize model evaluation metrics.

    Args:
        metrics (dict): A dictionary containing the metrics to plot.
        model_name (str): The name of the model.
    """

    metric_names = list(metrics.keys())
    metric_values = list(metrics.values())

    plt.figure(figsize=(8, 6))
    plt.bar(metric_names, metric_values, color=['blue', 'green', 'red', 'orange'])
    plt.title(f'Model Evaluation Metrics - {model_name}')
    plt.ylabel('Score')
    plt.ylim([0, 1])  # Set y-axis limits for better visualization
    plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability
    plt.tight_layout()
    plt.show()

# Assuming you have already calculated these metrics using:
# from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
lr_metrics = {
    "Accuracy": accuracy_score(y_test, lr_predictions),
    "Precision": precision_score(y_test, lr_predictions, zero_division=0),  # Handle zero division
    "Recall": recall_score(y_test, lr_predictions),
    "F1-Score": f1_score(y_test, lr_predictions),
}

plot_metrics(lr_metrics, "Logistic Regression")
```

```python
from IPython.display import HTML

def model_to_html(model, feature_names):
    """
    Generates HTML code for displaying model coefficients and intercept.

    Args:
        model (LogisticRegression): The Logistic Regression model.
        feature_names (list): List of feature names.
    """

    coefficients = model.coef_[0]
    intercept = model.intercept_[0]

    html_code = '<table><tr><th>Feature</th><th>Coefficient</th></tr>'
    for i, feature_name in enumerate(feature_names):
        html_code += f'<tr><td>{feature_name}</td><td>{coefficients[i]:.4f}</td></tr>'
    html_code += f'<tr><td>Intercept</td><td>{intercept:.4f}</td></tr></table>'

    return html_code

# Generate HTML code for Logistic Regression model
lr_model_html_code = model_to_html(lr_model, X_train.columns)  # Replace X_train.columns with your feature names if different

# Display the HTML code in Colab
display(HTML(lr_model_html_code))
```
